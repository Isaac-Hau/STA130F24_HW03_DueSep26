````python
import pandas as pd
import plotly.graph_objects as go
import numpy as np

# Load the dataset from the URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
penguins = pd.read_csv(url)

# Drop NaN values in flipper_length_mm and species columns
penguins = penguins.dropna(subset=['flipper_length_mm', 'species'])

# List of unique species
species_list = penguins['species'].unique()

# Create the figure
fig = go.Figure()

# Loop through each species to create separate histograms
for species in species_list:
    # Filter data for the species
    species_data = penguins[penguins['species'] == species]['flipper_length_mm']
    
    # Calculate statistics
    mean = species_data.mean()
    median = species_data.median()
    min_val = species_data.min()
    max_val = species_data.max()
    q1 = species_data.quantile(0.25)
    q3 = species_data.quantile(0.75)
    std = species_data.std()
    
    # Create the histogram
    fig.add_trace(go.Histogram(x=species_data, name=f'{species}', opacity=0.6))

    # Add vertical lines for mean and median
    fig.add_vline(x=mean, line_dash='dash', line_color='blue', annotation_text=f'{species} Mean', annotation_position="top left")
    fig.add_vline(x=median, line_dash='dot', line_color='green', annotation_text=f'{species} Median', annotation_position="top right")

    # Add vertical rectangles for range, IQR, and ±2 std deviations
    fig.add_vrect(x0=min_val, x1=max_val, fillcolor="lightblue", opacity=0.2, line_width=0, annotation_text=f'{species} Range')
    fig.add_vrect(x0=q1, x1=q3, fillcolor="orange", opacity=0.3, line_width=0, annotation_text=f'{species} IQR')
    fig.add_vrect(x0=mean - 2*std, x1=mean + 2*std, fillcolor="purple", opacity=0.1, line_width=0, annotation_text=f'{species} ±2 Std Dev')

# Update layout
fig.update_layout(title='Flipper Length Distribution with Statistical Markers for Each Species',
                  xaxis_title='Flipper Length (mm)',
                  yaxis_title='Count',
                  barmode='overlay')

# Show the figure
fig.show(renderer="png")
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the dataset from the URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
penguins = pd.read_csv(url)

# Drop NaN values in flipper_length_mm and species columns
penguins = penguins.dropna(subset=['flipper_length_mm', 'species'])

# List of unique species
species_list = penguins['species'].unique()

# Create a figure and axes
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Loop through each species and create the KDE plot
for i, species in enumerate(species_list):
    # Filter data for the species
    species_data = penguins[penguins['species'] == species]['flipper_length_mm']
    
    # Calculate statistics
    mean = species_data.mean()
    median = species_data.median()
    min_val = species_data.min()
    max_val = species_data.max()
    q1 = species_data.quantile(0.25)
    q3 = species_data.quantile(0.75)
    std = species_data.std()
    
    # Create the KDE plot
    sns.kdeplot(species_data, ax=axes[i], fill=True, color="skyblue")
    axes[i].set_title(f'{species}')
    axes[i].set_xlabel('Flipper Length (mm)')
    
    # Add vertical lines for mean and median
    axes[i].axvline(mean, color='blue', linestyle='--', label='Mean')
    axes[i].axvline(median, color='green', linestyle=':', label='Median')

    # Add shaded regions for range, IQR, and ±2 std deviations
    axes[i].axvspan(min_val, max_val, alpha=0.2, color='lightblue', label='Range')
    axes[i].axvspan(q1, q3, alpha=0.3, color='orange', label='IQR')
    axes[i].axvspan(mean - 2*std, mean + 2*std, alpha=0.1, color='purple', label='±2 Std Dev')

# Add a legend to the last plot
axes[-1].legend()

# Set the overall title
fig.suptitle('Flipper Length Distribution with KDE and Statistical Markers for Each Species', fontsize=16)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show the plot
plt.show()
````

Here’s a "pros and cons" list for the three data distribution visualization methods (Box Plot, Histogram, and KDE):

---

### **1. Box Plot**

#### **Pros**:
- **Simple and concise**: Provides a quick summary of the data using the five-number summary (minimum, Q1, median, Q3, maximum).
- **Outlier detection**: Clearly highlights outliers, which is useful for identifying unusual data points.
- **Compact**: Takes up minimal space, especially when comparing multiple distributions side-by-side.
- **Robust to skewness**: Shows asymmetry in the data by the position of the median relative to the quartiles.

#### **Cons**:
- **No distribution shape**: It doesn’t show the actual distribution shape (e.g., unimodal, bimodal, etc.).
- **No frequency insight**: Provides no direct insight into how frequently values occur within the range.
- **Less detailed**: Lacks granularity compared to histograms or KDE, making it harder to spot subtle patterns or nuances in the data.

---

### **2. Histogram**

#### **Pros**:
- **Distribution shape**: Offers a visual representation of the shape of the data distribution (unimodal, bimodal, skewed, etc.).
- **Easy to interpret**: Frequencies or counts in each bin are easy to understand.
- **Good for large datasets**: Effectively handles large amounts of data, allowing patterns to emerge clearly.

#### **Cons**:
- **Bin width sensitivity**: The choice of bin size can greatly affect the interpretation. Too few bins can oversimplify; too many can overcomplicate.
- **Discontinuous**: Shows a discrete view of the data, which may miss finer details between bin boundaries.
- **Arbitrary boundaries**: The boundaries of bins can obscure the real distribution if not chosen carefully.

---

### **3. Kernel Density Estimator (KDE)**

#### **Pros**:
- **Smooth visualization**: Provides a continuous and smoothed estimate of the distribution, giving a clearer sense of the data’s underlying density.
- **Handles multimodality**: Captures multiple modes or peaks in the data better than box plots or histograms.
- **Flexible**: Unlike histograms, KDEs do not rely on arbitrary bin boundaries, so they can provide a more natural estimate of the data's distribution.

#### **Cons**:
- **Bandwidth sensitivity**: The result is highly dependent on the bandwidth (smoothing) parameter, which can make the estimate too spiky or too smooth.
- **Less intuitive**: For some audiences, the probability density function (PDF) representation may be less intuitive than a simple histogram or box plot.
- **Data masking**: It can obscure actual data points, giving a more theoretical view rather than showing exact counts or specific data values.

---

### **Summary Comparison**:

- **Level of Detail**:
   - **Box Plot**: Low; focuses on summary statistics.
   - **Histogram**: Moderate; shows more details with binning.
   - **KDE**: High; gives the smoothest and most detailed visualization of the data.

- **Ease of Interpretation**:
   - **Box Plot**: High; quick and easy for basic data comparisons.
   - **Histogram**: High; frequencies and distribution shape are clear.
   - **KDE**: Medium; offers a more nuanced view but can be harder for non-experts to interpret.

- **Best for Identifying**:
   - **Box Plot**: Outliers and summary statistics.
   - **Histogram**: General shape of the distribution and approximate frequencies.
   - **KDE**: Fine details, multimodality, and overall distribution pattern.

---

Each method has its strengths and weaknesses, and they complement each other based on what you want to extract from the data—whether it’s simplicity, frequency detail, or subtle patterns.

Here’s a concise summary of the "pros and cons" of Box Plots, Histograms, and Kernel Density Estimators (KDE) for visualizing data distributions:

- **Box Plot**: Offers a simple, compact summary of key statistics (e.g., median, quartiles, outliers), but lacks details on the distribution's shape and frequency.
  
- **Histogram**: Shows the shape of the data distribution using bins, making it easy to interpret frequencies, but the choice of bin width can affect the clarity and accuracy of the visualization.
  
- **KDE**: Provides a smooth and continuous estimate of the data's distribution, capturing finer details and multiple peaks, though it requires careful bandwidth selection and may be less intuitive for some users.

In summary:
- Box plots are best for quick, high-level comparisons and identifying outliers.
- Histograms are more detailed but sensitive to binning.
- KDEs offer the most nuanced view of the distribution but require more careful tuning and interpretation.

````python
from scipy import stats
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

n = 1500
data1 = stats.uniform.rvs(0, 10, size=n)
data2 = stats.norm.rvs(5, 1.5, size=n)
data3 = np.r_[stats.norm.rvs(2, 0.25, size=int(n/2)), stats.norm.rvs(8, 0.5, size=int(n/2))]
data4 = stats.norm.rvs(6, 0.5, size=n)

fig = make_subplots(rows=1, cols=4)

fig.add_trace(go.Histogram(x=data1, name='A', nbinsx=30, marker=dict(line=dict(color='black', width=1))), row=1, col=1)
fig.add_trace(go.Histogram(x=data2, name='B', nbinsx=15, marker=dict(line=dict(color='black', width=1))), row=1, col=2)
fig.add_trace(go.Histogram(x=data3, name='C', nbinsx=45, marker=dict(line=dict(color='black', width=1))), row=1, col=3)
fig.add_trace(go.Histogram(x=data4, name='D', nbinsx=15, marker=dict(line=dict(color='black', width=1))), row=1, col=4)

fig.update_layout(height=300, width=750, title_text="Row of Histograms")
fig.update_xaxes(title_text="A", row=1, col=1)
fig.update_xaxes(title_text="B", row=1, col=2)
fig.update_xaxes(title_text="C", row=1, col=3)
fig.update_xaxes(title_text="D", row=1, col=4)
fig.update_xaxes(range=[-0.5, 10.5])

for trace in fig.data:
    trace.xbins = dict(start=0, end=10)
    
# This code was produced by just making requests to Microsoft Copilot
# https://github.com/pointOfive/stat130chat130/blob/main/CHATLOG/wk3/COP/SLS/0001_concise_makeAplotV1.md

fig.show(renderer="png") # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

   - The **mean** is the average of a dataset, while the **median** is the middle value when the data is ordered. The mean is sensitive to outliers, whereas the median is more robust against them.

2. **Skewness**:
   - **Right Skewness (Positive Skew)**: Most data points are clustered to the left with a long tail on the right. The mean is typically greater than the median due to high outliers pulling the mean upwards.
   - **Left Skewness (Negative Skew)**: Most data points are clustered to the right with a long tail on the left. The mean is typically less than the median due to low outliers pulling the mean downwards.

3. **Code Explanation**:
   - You shared a code snippet that generates two samples from a gamma distribution (right-skewed) and its negative (left-skewed). The code calculates the mean and median for both samples and plots their distributions.
   - The first sample (right-skewed) demonstrates that the mean is higher than the median, while the second sample (left-skewed) shows the opposite.

4. **Impact of Skewness on Mean and Median**:
   - The mean is influenced by all values in the dataset, making it sensitive to extreme values (outliers) in either direction, while the median remains stable as it only reflects the central position of the data.

This conversation focused on understanding how skewness affects measures of central tendency, providing clarity on the distinctions between mean and median in different data distributions.

````python
import plotly.graph_objects as go

import pandas as pd

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
dataset = pd.read_csv(url)

years = ["1952", "1962", "1967", "1972", "1977", "1982", "1987", "1992", "1997", "2002",
         "2007"]

# make list of continents
continents = []
for continent in dataset["continent"]:
    if continent not in continents:
        continents.append(continent)
# make figure
fig_dict = {
    "data": [],
    "layout": {},
    "frames": []
}

# fill in most of layout
fig_dict["layout"]["xaxis"] = {"range": [30, 85], "title": "Life Expectancy"}
fig_dict["layout"]["yaxis"] = {"title": "GDP per Capita", "type": "log"}
fig_dict["layout"]["hovermode"] = "closest"
fig_dict["layout"]["updatemenus"] = [
    {
        "buttons": [
            {
                "args": [None, {"frame": {"duration": 500, "redraw": False},
                                "fromcurrent": True, "transition": {"duration": 300,
                                                                    "easing": "quadratic-in-out"}}],
                "label": "Play",
                "method": "animate"
            },
            {
                "args": [[None], {"frame": {"duration": 0, "redraw": False},
                                  "mode": "immediate",
                                  "transition": {"duration": 0}}],
                "label": "Pause",
                "method": "animate"
            }
        ],
        "direction": "left",
        "pad": {"r": 10, "t": 87},
        "showactive": False,
        "type": "buttons",
        "x": 0.1,
        "xanchor": "right",
        "y": 0,
        "yanchor": "top"
    }
]

sliders_dict = {
    "active": 0,
    "yanchor": "top",
    "xanchor": "left",
    "currentvalue": {
        "font": {"size": 20},
        "prefix": "Year:",
        "visible": True,
        "xanchor": "right"
    },
    "transition": {"duration": 300, "easing": "cubic-in-out"},
    "pad": {"b": 10, "t": 50},
    "len": 0.9,
    "x": 0.1,
    "y": 0,
    "steps": []
}

# make data
year = 1952
for continent in continents:
    dataset_by_year = dataset[dataset["year"] == year]
    dataset_by_year_and_cont = dataset_by_year[
        dataset_by_year["continent"] == continent]

    data_dict = {
        "x": list(dataset_by_year_and_cont["lifeExp"]),
        "y": list(dataset_by_year_and_cont["gdpPercap"]),
        "mode": "markers",
        "text": list(dataset_by_year_and_cont["country"]),
        "marker": {
            "sizemode": "area",
            "sizeref": 200000,
            "size": list(dataset_by_year_and_cont["pop"])
        },
        "name": continent
    }
    fig_dict["data"].append(data_dict)

# make frames
for year in years:
    frame = {"data": [], "name": str(year)}
    for continent in continents:
        dataset_by_year = dataset[dataset["year"] == int(year)]
        dataset_by_year_and_cont = dataset_by_year[
            dataset_by_year["continent"] == continent]

        data_dict = {
            "x": list(dataset_by_year_and_cont["lifeExp"]),
            "y": list(dataset_by_year_and_cont["gdpPercap"]),
            "mode": "markers",
            "text": list(dataset_by_year_and_cont["country"]),
            "marker": {
                "sizemode": "area",
                "sizeref": 200000,
                "size": list(dataset_by_year_and_cont["pop"])
            },
            "name": continent
        }
        frame["data"].append(data_dict)

    fig_dict["frames"].append(frame)
    slider_step = {"args": [
        [year],
        {"frame": {"duration": 300, "redraw": False},
         "mode": "immediate",
         "transition": {"duration": 300}}
    ],
        "label": year,
        "method": "animate"}
    sliders_dict["steps"].append(slider_step)


fig_dict["layout"]["sliders"] = [sliders_dict]

fig = go.Figure(fig_dict)

fig.show(renderer="png")
````
